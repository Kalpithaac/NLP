```python
def extract_text_from_pdf(file_path):
    with open(file_path, 'rb') as file:
        reader = PyPDF2. PdfReader(file)
        text = ''
        for page in range(len(reader.pages)):
            page_content = reader.pages[page].extract_text()
            text += page_content
        return text
```

```python
def remove_spl_chars(text):
    text = re.sub('[^a-zA-Z0-9]', ' ', text)
    text = re.sub('\s+', ' ', text)
    return text
```

```python

def extract_keyphrases(doc):
    tokens = word_tokenize(doc)
    stop_words = set(stopwords.words('english'))
    lemmatizer = WordNetLemmatizer()
    filtered_tokens = [lemmatizer.lemmatize(token) for token in tokens if token.isalnum() and token not in stop_words]
    filtered_document = ' '.join(filtered_tokens)
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform([filtered_document])

    # Get the feature names (keyphrases)
    feature_names = vectorizer.get_feature_names_out()
    


    n = 15
    top_keyphrases = [feature_names[index] for index in tfidf_matrix.indices[np.argsort(tfidf_matrix.data)[-n:]]]

    return top_keyphrases
```

```python
top_keyphrases = extract_keyphrases(doc)
```

```python
print("Top keyphrases:")
for keyphrase in top_keyphrases:
    print(keyphrase)
```
